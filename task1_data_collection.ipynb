```python
#import library
import requests
import time
from datetime import datetime
import os
import json
import re

# call api and store ids
response = requests.get("https://hacker-news.firebaseio.com/v0/topstories.json/")
try:
    top_stories = response.json()
except Exception as e:
    print("Error:", e)
```

```python
# i extended few key words in worldnews, sports and science as i got only 78 matches 
category = {
            'technology' : ['AI', 'software', 'tech', 'code', 'computer', 'data', 'cloud', 'API', 'GPU', 'LLM'],
            'worldnews' : ['war', 'government', 'country', 'president', 'election', 'climate', 'attack', 'global','china','russia','ukraine','pentagon','nato','trump','iran','white house','nato'],
            'sports' : ['NFL', 'NBA', 'FIFA', 'sport', 'game', 'team', 'player', 'league', 'championship','tournament','olympic','stadium','esport','cricket','athlete','racing','chess'],
            'science' : ["research", "study", "space", "physics", "biology", "discovery", "NASA", "genome",'radioactive','asteroid','scientist','artemis','quantum','neuroscience'],
            'entertainment' : ["movie", "film", "music", "Netflix", "game", "book", "show", "award", "streaming",'anime','youtube','concert'],
           }
story_datas = []
total_story = 0

for key,values in category.items():
    count = 0
    #call each story and i use simple white space split for matching keyword
    for story_id in top_stories:
        try:
            story = requests.get(headers={"User-Agent": "TrendPulse/1.0"}, url=f"https://hacker-news.firebaseio.com/v0/item/{story_id}.json")
            if story.status_code != 200:
                continue
            story_response = story.json()
            title = story_response['title'].lower().split()
            if any(kw.lower() in title for kw in values): 
                print(title)
                # max cap on each category
                if count >= 25:
                    break
                count  += 1
                print(count)
                # add all the required fields here
                data = {
                    'post_id' : story_response['id'],
                    'title' : story_response['title'],
                    'category' : key,
                    'score' : story_response['score'],
                    'num_comments' : story_response['descendants'],
                    'author' : story_response['by'],
                    'collected_at' : str(datetime.now()),
                    }
                story_datas.append(data)
                
            
        except Exception as e:
            print("Error:", e)
    # time gap for next category
    time.sleep(2)
        
```

```python
# check and create folder if not exist
os.makedirs("data", exist_ok=True)
date_str = datetime.now().strftime("%Y%m%d")
file_path = f"data/trends_{date_str}.json"
with open(file_path, "w", encoding="utf-8") as f:
    json.dump(story_datas, f)
# total story added in json
print(f"Total stories collected: {len(story_datas)}")
print(f"Saved to: {file_path}")
```
